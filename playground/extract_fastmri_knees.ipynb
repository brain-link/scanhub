```python
# IMPORTS
import numpy as np
import h5py
from glob import glob
import matplotlib.pyplot as plt
```

```python
# Directory with hdf5 fast mri data files
data_path = "/Users/davidschote/Projects/data/fastmri_knee/"
data_files = glob(f"{data_path}*.h5")

# Read a single sample
data = h5py.File(data_files[0], mode="r")
print(data.keys())
ksp = data["kspace"]
print(ksp.dtype)
print(ksp.shape)
```

```python
# Define the reconstruction
def ksp_to_img(x):
    return np.array([np.fft.fftshift(np.fft.fft2(np.fft.fftshift(x[k, ...]))) for k in range(x.shape[0])])

# Show a single sample slice
img = ksp_to_img(ksp)
plt.imshow(np.abs(img[10]))
```

```python
# Reconstruct all h5 files and store reconstruction result with numpy
target_path = "/Users/davidschote/Projects/data/fastmri_reco_numpy/"

for k, file in enumerate(data_files):
    data = h5py.File(data_files[0], mode="r")
    img = ksp_to_img(data["kspace"])
    np.save(f"{target_path}img_{k}.npy", img)
    print(f"\rProgress: {round((k+1)*100/len(data_files), 1)}%")
```
